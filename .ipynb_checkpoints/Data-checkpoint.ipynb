```python
# Tratando dados ausentes
def trataDados(data):

    bodyMediana = data.body.mean()
    data.fillna(bodyMediana, inplace = True)
    ageMediana = data.age.mean()
    data.fillna(ageMediana, inplace=True)
    data['sex'] = data['sex'].map({'female': 0, 'male': 1}).astype(int)
    
    print("Valores NaN: ", data.isnull().sum().sort_values(ascending = False)
```

```python
def Acuracia(y_test, pred):
    print("Matriz de confusão: ",metrics.confusion_matrix(y_test, pred))
    print("Acuracia: ",metrics.accuracy_score(y_test, pred))
```

```python
def Arvore(clf, X, Y):
    print("Representação textual da árvore: ", tree.export_text(clf))
    fig = plt.figure(figsize=(25,25)) 
    _= tree.plot_tree(clf,
                        feature_names = X,
                        class_names=Y,
                        filled = True)
```
